#1. Análisis de facturación



```python
#Facturacion por separado tienda1
FacturacionT1= sum(tienda['Precio'])
print(f'Facturacion total Tienda1 ${FacturacionT1}')
```

```python
#Facturacion por separado tienda2
FacturacionT2= sum(tienda2['Precio'])
print(f'Facturacion total Tienda2 ${FacturacionT2}')
```

```python
#Facturacion por separado tienda3
FacturacionT3= sum(tienda3['Precio'])
print(f'Facturacion total Tienda3 ${FacturacionT3}')
```

```python
#Facturacion por separado tienda4
FacturacionT4 = sum(tienda4['Precio'])
print(f'Facturacion total Tienda4 ${FacturacionT4}')
```

```python
#calculo de Facturacion con ciclo (for) implementando una lista de tiendas
# Lista de tiendas
tiendas = {"Tienda1": tienda, "Tienda2": tienda2, "Tienda3": tienda3, "Tienda4": tienda4}

# Calcular y mostrar facturación total por tienda
for nombre, datos in tiendas.items():
    facturacion = sum(datos["Precio"])
    print(f'Facturación total {nombre}: ${facturacion}')
```

```python
import matplotlib.pyplot as plt

# Crear lista de nombres y valores
nombres_tiendas = list(tiendas.keys())
facturacion_tiendas = [sum(datos["Precio"]) for datos in tiendas.values()]

# Graficar la facturación por tienda
plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas, facturacion_tiendas, color='green')
plt.xlabel("Tiendas")
plt.ylabel("Facturación ($)")
plt.title("Facturación por Tienda")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

# 2. Ventas por categoría

```python
#Analisis de productos vendidos por categoria en la tienda 1

# Crear un diccionario para almacenar el conteo de productos por categoría en Tienda 1
conteo_categoria_tienda1 = {}

# Recorrer cada fila del DataFrame de Tienda 1
for index, row in tienda.iterrows():
    categoria = row["Categoría del Producto"]

    # Si la categoría ya está en el diccionario, incrementar el conteo
    if categoria in conteo_categoria_tienda1:
        conteo_categoria_tienda1[categoria] += 1
    else:
        conteo_categoria_tienda1[categoria] = 1

# Mostrar los datos procesados
print("\nCategorías más vendidas en Tienda 1:")
for categoria, cantidad in conteo_categoria_tienda1.items():
    print(f"{categoria}: {cantidad}")
```

```python

categorias = list(conteo_categoria_tienda1.keys())
cantidades = list(conteo_categoria_tienda1.values())

plt.figure(figsize=(8, 5))
plt.bar(categorias, cantidades, color='skyblue')
plt.xlabel("Categoría")
plt.ylabel("Cantidad Vendida")
plt.title("Cantidad de productos vendidos por categoría - Tienda 1")
plt.xticks(rotation=45)  # Rotar etiquetas para mejor lectura
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
#Analisis de productos vendidos por categoria en la tienda 2

#Creamos un diccionario para almacenar los productos por categoria en Tienda 2
conteo_categoria_tienda2={}

#recorremos cada fila de tienda2
for index, row in tienda2.iterrows():
  categoria= row["Categoría del Producto"]
  #Corroboramos la categoria si ya pertenece al diccionario de conteo se incrementa el conteo (productos Vendidos)
  if categoria in conteo_categoria_tienda2:
      conteo_categoria_tienda2[categoria]+=1
  else:
      conteo_categoria_tienda2[categoria]=1

#Mostramos los datos obtenidos
print("\nCategorias mas vendidas en Tienda 2:")
for categoria, cantidad in conteo_categoria_tienda2.items():
  print(f"{categoria}: {cantidad}")
```

```python

categorias = list(conteo_categoria_tienda2.keys())
cantidades = list(conteo_categoria_tienda2.values())

plt.figure(figsize=(8, 5))
plt.bar(categorias, cantidades, color='skyblue')
plt.xlabel("Categoría")
plt.ylabel("Cantidad Vendida")
plt.title("Cantidad de productos vendidos por categoría - Tienda 2")
plt.xticks(rotation=45)  # Rotar etiquetas para mejor lectura
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
